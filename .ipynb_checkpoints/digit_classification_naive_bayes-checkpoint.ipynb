### Applying a smoothing using numpy.
---

A common image processing technique is to smooth an image by **blurring**. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian, i.e., the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

1. Implement a simplified Gaussian blur filter by just using the 8 neighboring pixels like this: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values.
1. Choose some weights as a starting point. Produce and evaluate four 1-Nearest Neighbor models by applying your blur filter in these ways:
  1. Do not use the filter
  1. Filter the training data but not the dev data
  1. Filter the dev data but not the training data
  1. Filter both training data and dev data
1. Show the accuracies of the four models evaluated as described. What do you see? Can you explain why this is?
1. Experiment with weights that makes one model's accuracy at least 0.9.

```python
def blur(data, kernel):
    new_data = np.zeros((1000,784))
    for i in range(0,len(data)):
        old = data[i].reshape(28,28)
        #pad incoming data to make it 30x30 so we don't have to handle edge cases
        old_padded = np.pad(old, pad_width=1, mode='constant', constant_values=0)
        new = new_data[i].reshape(28,28)
        for r in range(1,28):
            for c in range(1,28):
                #take the sum of all the weights multipled by their corresponding existing value
                new[r-1][c-1] = np.sum(old_padded[r-1:r+2, c-1:c+2] * kernel)              
        #reshape back to vector form
        new_data[i] = new.reshape(784,)
    return new_data     
```

Here you'll likely need to implement a `trianarize` helper function. Since we are going beyond the binary representation of our features, you should use `MultinomialNB` to produce the model.

```python
#binarize data based on threshold of 0.1
def binarize(data):
    bi_data = np.zeros(data.shape)
    for i in range(0,len(data)):
        if data[i] >= 0.1:
            bi_data[i] = 1
        else:
            bi_data[i] = 0 
    return bi_data

#pixel < 0.1 = 0
#0.1 < pixel < 0.9 = 1
#else, 2
def trinarize(data):
    tri_data = np.zeros(data.shape)
    for i in range(0,len(data)):
        if data[i] < 0.1:
            tri_data[i] = 0
        elif data[i] >= 0.1 and data[i] < 0.9:
            tri_data[i] = 1
        else:
            tri_data[i] = 2
    return tri_data
```

```python
bi_mini = np.apply_along_axis(func1d = binarize, axis = 1, arr = mini_train_data)
tri_mini = np.apply_along_axis(func1d = trinarize, axis = 1, arr = mini_train_data)
bi_dev = np.apply_along_axis(func1d = binarize, axis = 1, arr = dev_data)
tri_dev = np.apply_along_axis(func1d = trinarize, axis = 1, arr = dev_data)
```

```python
def Q2():
    #create bernoulli model and train on binarized data that was declared outside of funciton
    bnb = BernoulliNB()
    bnb.fit(bi_mini, mini_train_labels)
    print("The Bernoulli model's accuracy is {:.3f}".format(bnb.score(bi_dev, dev_labels)))
    
    #Model 2 - Trinarize
    mnb = MultinomialNB()
    mnb.fit(tri_mini, mini_train_labels)
    print("The Multinomial model's accuracy is {:.3f}".format(mnb.score(tri_dev, dev_labels)))
Q2()
```

The multinomial model's accuracy is slightly lower than the Bernoulli. This result makes sense, because we set very wide trinarizing thresholds of 0.1 and 0.9, meaning a lot of the data was being greyed out which just adds noise to a previously fairly binarized photo.

### Applying the grid search technique.
1. Search across several values of the LaPlace smoothing parameter (alpha) to find its effect on a Bernoulli Naive Bayes model's performance.  Show the accuracy at each alpha value.
1. What is the best value for alpha? What is the accuracy when alpha is near 0? Is this what you'd expect?

```python
def Q3(alphas):

### STUDENT START ###
    #grid search through the inputted values of alpha
    grid = GridSearchCV(estimator = BernoulliNB(binarize = 0.0), param_grid = alphas, cv=5, scoring='accuracy')
    
    grid.fit(bi_mini, mini_train_labels)
    #get the scores from cv_results for each alpha
    scores = grid.cv_results_['mean_test_score']
    
    #print out accuracies
    for i in range(len(scores)):
        print('For alpha = {}, the accuracy was {:.4f}'.format(alphas['alpha'][i],round(scores[i], 4)))
### STUDENT END ###

alphas = {'alpha': [1.0e-10, 0.0001, 0.001, 0.01, 0.1, 0.5, 1.0, 2.0, 10.0]}
Q3(alphas)
```

The model accuracies are relatively similar for each of the various alpha values, however the highest accuracy was achieved with an alpha value = 0.01. It makes sense that the accuracy declines as we increase alpha substantially because it causes the likelihood probability to move towards a uniform distribution (0.5). With pixel values all within a smaller range, it will make it harder to distinguish between digits.

```python
def Q4():

### STUDENT END ###
    gnb = GaussianNB()
    #fit on non-binarized data 
    gnb.fit(mini_train_data, mini_train_labels)
    print("For the Gaussian Naive Bayes model, the accuracy is: {:.3f}".format(gnb.score(dev_data, dev_labels)))
    
    #mean of each feature per class
    print("The mean of theta_ for the Gaussian Model as is: ", np.mean(gnb.theta_))
    #variance of each feature per class
    print("The mean of sigma_ for the Gaussian Model as is: ", np.mean(gnb.sigma_))
    
    #after seeing the theta_ and sigma_ values, I wanted to test values around ~0.05-0.06
    tr = [0.05, 0.055, 0.06, 0.065, 0.075, 0.08, 0.085, 0.09, 2]
    #improved gaussian model
    for t in tr:
        gnb_t = GaussianNB(var_smoothing = t)
        gnb_t.fit(mini_train_data, mini_train_labels)
        print("After changing the value of var_smoothing to {}".format(t))
        print("The accuracy of the improved Gaussian Naive Bayes model is: {:.3f}".format(gnb_t.score(dev_data, dev_labels))) 

### STUDENT END ###

Q4()
```

ANSWER: When a Gaussian Naive Bayes model is implemented, it essentially gives more weights to the samples that are closer to the distribution mean. So, this model needed some sort of adjustment that allowed more samples into the mix. What the var_smoothing parameter allows us to do is add a user-defined value to the distribution's variance so that it will account for more samples that are further away from the distribution mean. From pulling the mean values of theta_ and sigma_ values of the original GNB model, we see that the mean of the variances is 0.05, so I wanted to test var_smoothing values around that number and test their accuracies. As shown in the print statement, the var_smoothing value that led to the highest accuracy was 0.06. 

### Naive Bayes as a generative model
---
Because Naive Bayes produces a generative model, you can use it to generate digit images.

1. Produce a Bernoulli Naive Bayes model and then use it to generate a 10x10 grid with 10 example images of each digit. Each pixel output will be either 0 or 1: randomly generating a number and then comparing it to the estimated probability of the pixel being either 0 or 1.  Show the grid.
1. How do the generated digit images compare to the training digit images?

```python
def Q5(num_examples):

### STUDENT START ###
    nb = BernoulliNB(alpha = 0.01) 
    nb.fit(bi_mini, mini_train_labels) 
    probs = np.exp(nb.feature_log_prob_)
    
    # plot the images
    fig, ax = plt.subplots(10,10, figsize = (10,10))
    for i in range(0, 10):
        for j in range(0, 10):
            #generate 784 random numbers
            rand_nums = np.random.rand(784)
            pixels = np.zeros(784)
            for k in range(784):
                #only make a pixel val 1 if the random num is less than
                #the log prob of the feature for that class
                if rand_nums[k] <= probs[j][k]:
                    pixels[k] = 1
            #reshape and show
            pic = pixels.reshape(28,28)
            ax[j,i].imshow(pic, cmap = 'binary')
            ax[j,i].set_xticks([])
            ax[j,i].set_yticks([]) 
### STUDENT END ###

Q5(10)
```

ANSWER: The generated digits are much more scattered and harder to make out than the images in the training data. This is because for our initial Bernoulli NB model, we were binarizing each pixel on a threshold of 0.1 and not taking any other pixels from other images as input. With this problem, we're taking P(Xi | Y) and seeing if it had a higher probability than a random number that was generated. If there was a greater variance in that feature, it's likely that it could be categorized as the wrong binary. That's what leads to the randomness present in the images of these digits.

### Model Calibraiton
---
Recall that:
 * a **strongly calibrated** classifier is rougly 90% accurate when it says it is 90% accurate. The model's estimated posterior probability of the predicted class is indeed 0.9. 
 * A **weakly calibrated** classifier is more accurate when it "says" it is more accurate. The actual accuracy is higher when the model's estimated posterior probability is higher.
 * A **poorly calibrated** classifier has no positive correlation between the model's estimate posterior probability and the actual accuracy.

1. Produce a Bernoulli Naive Bayes model.  
1. Evaluate performance: 
  1. Partition the dev set into several buckets based on the estimated posterior probabilities of the predicted classes (predict_proba)
    - Think of it as a bin in a histogram, where each bin groups a range of estimated posterior probabilities of the predicted classes (predict_proba).
    - Then estimate the actual accuracy the classifier achieved for each bucket. 
    - So, for each prediction:
       - Find the bucket whose range includes the estimated posterior probability, 
       - and update "correct" and "total" counters accordingly. 
       - Show the accuracy for each bucket.
1. How would you characterize the calibration for this Bernoulli Naive Bayes model according to the definitions above?

```python
def Q6(buckets, correct, total):
    
### STUDENT START ###
#Step 1
    #(a) Fit a Bernoulli model on the train data
    nb = BernoulliNB(alpha = 0.01, binarize=0.0) # alpha set to grid search's best result
    nb.fit(mini_train_data, mini_train_labels) 
    #(b) Predict labels using the test data
    pred_vals = nb.predict(dev_data)
    #(c) Predict probabilities using the test data
    pred_probs = nb.predict_proba(dev_data)
#Step 2: (notice the nested loop here)
    #For each row in the test data:
    for i in range(0,len(dev_data)):
        #(a) extract predicted label
        p_lab = pred_vals[i]
        #(b) extract predicted probability
        p_pr = np.max(pred_probs[i])
        #(c) extract true label
        tr_lab = dev_labels[i]
        #For each bucket in buckets:
        for b in range(0,len(buckets)):
            #(a) if predicted probability <= bucket:
            if p_pr <= buckets[b]:
                #(a.1) add 1 to total corresponding to the bucket
                total[b] += 1
                #(a.2) if predicted label == true label:
                if p_lab == tr_lab:
                    #(a.2.1) add 1 to correct corresponding to the bucket
                    correct[b] += 1
#Step 3: compute accuracy for each bucket (if total corresponding to the bucket > 0; accuracy = correct/total)
    for f in range(0,len(buckets)):
        accuracy = 0
        if total[f] > 0:
            accuracy = correct[f] / total[f]
        print("For bucket {}, the accuracy is {:.4f}".format(buckets[f], accuracy))
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

Q6(buckets, correct, total)
```

Given the definitions provided, this appears to be a poorly calibrated classifier, because the actual accuracy is much less than the estimated posterior probability. For example, when the posterior probability of the predicted class was 0.9, the accuracy was only 33%.
